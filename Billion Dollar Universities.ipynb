```python
# imports; don't touch!
%matplotlib inline

from bs4 import BeautifulSoup
import urllib2
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Originally, I saw the data for this in a Wall Street Journal article that cited Moody's.
# It was an image and it was pixelated. WSJ can do better.
# I scraped wikipedia for state budget numbers using BeautifulSoup.

wiki = 'http://en.wikipedia.org/wiki/List_of_U.S._state_budgets'
soup = BeautifulSoup(urllib2.urlopen(urllib2.Request(wiki)))

s = []
e = []

for row in soup.findAll("tr"):
    cells = row.findAll("td")
    # this makes sure I only get the table of length four
    if len(cells) == 4:
        s.append(cells[0].findAll(text = True))
        e.append(cells[1].findAll(text = True))

        
# and here's some magic to get rid of weird index errors
# and tweak how unicode shows up
state = []
for i, c in enumerate(s):
    state.append(s[i][1])

billions = []
for i in e:
    billions.append(float(i[0]))
```

```python
# I also scraped wikipedia for universities with over a billion in endowments.
# This section is functionally the same as the previous.

wiki2 = 'http://en.wikipedia.org/wiki/List_of_colleges_and_universities_in_the_United_States_by_endowment'
soup2 = BeautifulSoup(urllib2.urlopen(urllib2.Request(wiki2)))

u = []
b = []

for row in soup2.findAll("tr"):
    cells = row.findAll("td")
    if len(cells) == 11:
        u.append(cells[0].findAll(text = True))
        b.append(cells[1].findAll(text = True))

uni = []
for i, c in enumerate(u):
    uni.append(u[i][0])

cash = []
for i in b:
    cash.append(float(i[0][1:]))
```

```python
# Unfortunately, the wikitable didn't have locations on there, so I had to scrape separately.

urls = []

for row in soup2.findAll("tr"):
    cells = row.findAll("td")
    if len(cells) == 11:    
        for link in cells[0].find_all('a'):
            urls.append(link.get('href'))

url_list = []
for link in urls:
    if link[:5] == '/wiki':
        url_list.append(link)
        
place = []

for url in url_list:
    wiki = "http://en.wikipedia.org"
    location_soup = BeautifulSoup(urllib2.urlopen(urllib2.Request(wiki+url)))
    sidebar = location_soup.find("table", {"class": "infobox vcard" })
    location = sidebar.find("span", {"class" : "locality"})
    place.append(location)
```

```python
# Now that that's done, let's build some dataframes!
# pause the video while I clean this thing up by hand. There's some 'None's
# because apparently wikipedia isn't consistent across their infobox-vcards.
# So silly.

state_budget = pd.DataFrame(data=[state, billions]).transpose()
state_budget.columns=['state', 'budget']
#state_budget.to_csv("state.csv", encoding='utf-8')

uni_pre_csv = pd.DataFrame(data=[uni, cash, place]).transpose()
uni_pre_csv.columns=['university', 'endowment in billions', 'location']
#uni_pre_csv.to_csv("uni.csv", encoding='utf-8')
```

```python
# ...and unpause to read in the csvs.

DF1 = pd.read_csv("uni.csv")
DF1 = DF1.drop(DF1.columns[0], axis=1)

DF2 = pd.read_csv("state.csv")
DF2 = DF2.drop(DF2.columns[0], axis=1)
```

```python
# So what am I trying to figure out?
# One simple question: 
# Are there any places where the university endowment is greater than the state budget,
# and if so, what state(s)?

# Here's some more low-key panda-ing to clean up everything for a merge.
DF_college = DF1.groupby(["short_state"]).sum()
DF_states = DF2.groupby(["short_state"]).sum()

# Now, the thing to note here is that there's not a billion-dollar university in every state,
# which is why we do this:

DF_merged = DF_states.join(DF_college)
DF_merged = DF_merged.dropna()

DF_merged[DF_merged['budget in billions'] < DF_merged['endowment in billions']]

# The list here is really spare. CT and MA.
# I could have expected it given the schools that are there (Harvard, Yale).
```

```python
# This part is going to get a bit more complicated.
# First, we'll generate an endowment to budget ratio like so:

DF_merged['ratio'] = DF_merged['endowment in billions'] / DF_merged['budget in billions']

DF_merged.head()

# Looking good!
```

```python
# And then we'll select the states which have a ratio higher than .5

DF_merged[DF_merged['ratio'] >= .5]

# .5 is a magic number; I figure it's pretty shocking to have
# an organization in your state with almost half the state's budget
# going toward its own politics, police and enjoyment or, in the case of my
# alma mater, its own zip code.
```

```python
# Anyway, that's enough of that...

DF_merged.drop(['ratio'],inplace=True,axis=1)
```

```python
plot = DF_merged.plot(kind='bar', figsize=(12, 10))
```

```python
# The most shocking thing to me, seeing this graph, 
# is that there are several univerities with budgets greater than states.
# Time to move onto something a little different, where we can directly compare states and universities.

second_merge = pd.merge(DF1,DF2)
second_merge.drop(['state', 'city'],inplace=True,axis=1)
second_merge['percentage of state budget'] = second_merge['endowment in billions'] / second_merge['budget in billions']
second_merge.describe()
```

```python
# So here's some fascinating stuff.
# There's obviously 84 entries, but the perecntiles are not what I expected necessarily. 
# The outliers here for here aren't major outliers; about 25% of the data are greater than the mean.

len(second_merge[second_merge['endowment in billions'] >= 4.46])

# However, there's fewer that are one standard deviation away. Only seven. And they are:

second_merge[second_merge['endowment in billions'] >= 4.46 + 6.098]

# Texas and Massachusetts are representing here, but Connecticut is pretty well-endowed
# (not that we didn't know that already).
```
